```python
X = df.drop('WAGE', axis=1)
Y = df['WAGE']

x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.3)
```

```python
print(x_train.shape)
print(x_test.shape)

print(y_train.shape)
print(y_test.shape)
```

```python
rfr = RandomForestRegressor(max_depth=6, n_estimators=400)
rfr.fit(x_train, y_train)
```

```python
# rfr = RandomForestRegressor(max_depth=6, n_estimators=400)


rfr.fit(x_train, y_train)

y_pred_train = rfr.predict(x_train)
y_pred_test = rfr.predict(x_test)

print('====================================================')

print('cross_val')
kfold = KFold(n_splits=5)
cross_val = cross_val_score(rfr, x_train, y_train, cv=kfold)
print(cross_val)
print(cross_val.mean())

print('====================================================')
print('r2_score')
print(r2_score(y_train, y_pred_train))
print(r2_score(y_test, y_pred_test))

print('====================================================')
result = pd.DataFrame({'real' : y_test, 'pred' : y_pred_test})
result['diff'] = result['real'] - result['pred']
sns.scatterplot(data=result, x='real' , y='pred')
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()], color='red', linestyle='--')
plt.show()

top20 = pd.Series(rfr.feature_importances_, x_train.columns)
top20 = top20.sort_values(ascending=False)
print(top20)
sns.barplot(x=top20, y=top20.index)
plt.show()
```

```python
def model_check(model):
    model.fit(x_train, y_train)

    y_pred_train = model.predict(x_train)
    y_pred_test = model.predict(x_test)

    print('====================================================')

    print('cross_val')
    kfold = KFold(n_splits=5)
    cross_val = cross_val_score(model, x_train, y_train, cv=kfold)
    print(cross_val)
    print(cross_val.mean())

    print('====================================================')
    print('r2_score')
    print(r2_score(y_train, y_pred_train))
    print(r2_score(y_test, y_pred_test))
    
    print('====================================================')
    print(mean_absolute_error(y_test, y_pred_test))
    print(np.sqrt(mean_squared_error(y_test, y_pred_test)))


    sns.scatterplot(data=result, x='real' , y='pred', color='blue')  # ëª¨ë¸ì˜ ì˜ˆì¸¡ì¹˜ëŠ” íŒŒëž€ìƒ‰ìœ¼ë¡œ í‘œì‹œ
    plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()], color='red', linestyle='--')  # í‰ê·  ë¼ì¸
    plt.axhline(y_test.mean(), color='green')  # í‰ê·  ë¼ì¸ ì¶”ê°€ (ì´ˆë¡ìƒ‰)
    plt.xlabel('Real')  # xì¶• ë ˆì´ë¸”
    plt.ylabel('Predicted')  # yì¶• ë ˆì´ë¸”

    # ìž”ì°¨ í”Œë¡¯ ì¶”ê°€
    sns.residplot(data=result, x='real', y='diff', color='purple', scatter_kws={"s": 20, "alpha": 0.5})  
    # ìž”ì°¨ í”Œë¡¯ì„ í†µí•´ ì˜ˆì¸¡ê°’ê³¼ ì‹¤ì œê°’ ì‚¬ì´ì˜ ì°¨ì´ë¥¼ ë³´ì—¬ì¤ë‹ˆë‹¤. ìž”ì°¨ê°€ 0ì„ ì¤‘ì‹¬ìœ¼ë¡œ ë¶„í¬ë ìˆ˜ë¡ ì¢‹ìŠµë‹ˆë‹¤.

    plt.show()
    
    top20 = pd.Series(model.feature_importances_, x_train.columns)
    top20 = top20.sort_values(ascending=False)
    print(top20)
    sns.barplot(x=top20, y=top20.index)
    plt.show()
```

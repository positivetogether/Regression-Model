```python
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score, KFold
import shap
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error


df=pd.read_csv("chapter17_offline_mart.csv")
X = df.drop(['Date', 'Weekly_Sales','Store', 'Unemployment'], axis=1)
Y  =df['Weekly_Sales']

x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.3)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)

y_pred_train = lr.predict(x_train)
y_pred_test = lr.predict(x_test)

print('r2')
print(r2_score(y_train, y_pred_train))
print(r2_score(y_test, y_pred_test))

print('mae')
print(mean_absolute_error(y_train, y_pred_train))
print(mean_absolute_error(y_test, y_pred_test))

print('mse')
print(np.sqrt(mean_squared_error(y_train, y_pred_train)))
print(np.sqrt(mean_squared_error(y_test, y_pred_test)))

print('coef')
print(pd.DataFrame({'features' : x_train.columns, 'coef': lr.coef_}).sort_values(by='coef', ascending=False))


result = pd.DataFrame({'real': y_test, 'predicted': y_pred_test})
sns.scatterplot(x=result['real'],y=result['predicted'])
plt.plot([result['real'].min(),result['real'].max()], [result['real'].min(),result['real'].max()], color='red')
plt.axhline(result['real'].mean(), linestyle='--', color='blue')
```

```python
def  model_check(model):
        
        model.fit(x_train, y_train)

        y_pred_train = model.predict(x_train)
        y_pred_test = model.predict(x_test)

        print('r2')
        print(r2_score(y_train, y_pred_train))
        print(r2_score(y_test, y_pred_test))

        print('mae')
        print(mean_absolute_error(y_train, y_pred_train))
        print(mean_absolute_error(y_test, y_pred_test))

        print('mse')
        print(np.sqrt(mean_squared_error(y_train, y_pred_train)))
        print(np.sqrt(mean_squared_error(y_test, y_pred_test)))
        if model == lr:
            print('coef')
            print(pd.DataFrame({'features' : x_train.columns, 'coef': lr.coef_}).sort_values(by='coef', ascending=False))
        

        result = pd.DataFrame({'real': y_test, 'predicted': y_pred_test})
        sns.scatterplot(x=result['real'],y=result['predicted'])
        plt.plot([result['real'].min(),result['real'].max()], [result['real'].min(),result['real'].max()], color='red')
        plt.axhline(result['real'].mean(), linestyle='--', color='blue')
        plt.show()
        
        top20 = pd.Series(model.feature_importances_, x_train.columns)
        top20=top20.sort_values(ascending=False)
        print(top20.sort_values(ascending=False))
        
        sns.barplot(y=top20.index , x=top20)
        plt.show()
```
